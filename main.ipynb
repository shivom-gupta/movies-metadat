```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
%matplotlib inline
```

```python
df= pd.read_csv('movies-metadata.csv')
df.head()
```

```python
df.columns
```

```python
columns = ['id','title','budget', 'genres','original_language','popularity', 'production_companies', 'production_countries',
           'release_date', 'revenue', 'runtime','spoken_languages','vote_average', 'vote_count']
```

```python
df=df[columns]
```

```python
df.head()
```

```python
df.info()
```

```python
df.budget = pd.to_numeric(df['budget'], errors= 'coerce')
```

```python
df.release_date = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
df.popularity= pd.to_numeric(df['popularity'], errors= 'coerce')
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
import missingno
missingno.matrix(df)
```

```python
df.revenue.replace(np.nan, 0)
```

```python
df['spoken_languages'] = df['spoken_languages'].fillna('[]').apply(ast.literal_eval)
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df['production_companies'] = df['production_companies'].fillna('[]').apply(ast.literal_eval)
df['production_companies'] = df['production_companies'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df['production_countries'] = df['production_countries'].fillna('[]').apply(ast.literal_eval)
df['production_countries'] = df['production_countries'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df['genres'] = df['genres'].fillna('[]').apply(ast.literal_eval)
df['genres'] = df['genres'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df.head()
```

```python
df.dtypes
```

df.to_excel('cleaned.xlsx')

```python
sns.regplot(data=df, x='budget', y='revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Revenue VS Budget');
```

```python
df['month'] = df['release_date'].dt.month_name()
df.month = pd.Categorical(df['month'], categories=['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September','October', 'November', 'December'],ordered=True)
```

```python
sns.lineplot(data=df, x='month', y='revenue')
plt.title('Month vs Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=30);
```

```python
df['day']=df['release_date'].dt.day_name()
df['day']= pd.Categorical(df['day'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
```

```python
sns.lineplot(data=df, x='day', y='revenue')
plt.xlabel('Day of week')
plt.ylabel('Revenue')
plt.title('Day of the week VS Revenue')
plt.xticks(rotation=30);
```

```python
df.genres = df.genres.astype(str)
```

```python
y=df.groupby('genres')['revenue'].mean().sort_values(ascending=False)
y=pd.DataFrame(y)
y.reset_index(inplace=True)
y=y.head(10)
```

```python
sns.barplot(data=y, x='revenue', y='genres',color=sns.color_palette()[0])
plt.xlabel('Revenue')
plt.ylabel('Genres')
plt.title('Mean Revenue VS Genres');
```

```python
sns.regplot(data=df, x='popularity', y='revenue')
plt.xlabel('Popularity')
plt.ylabel('Revenue')
plt.title('Revenue VS Popularity');
```

```python
sns.regplot(data=df, x='vote_count', y='revenue')
plt.title('Number of votes VS Revenue')
plt.xlabel('Number of Votes')
plt.ylabel('Revenue');
```

```python
sns.regplot(data=df, x='vote_average', y='revenue')
plt.title('Average score from votes VS Revenue')
plt.xlabel('Average score from votes')
plt.ylabel('Revenue');
```
